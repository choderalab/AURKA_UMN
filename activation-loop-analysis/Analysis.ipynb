```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.load('data/data.npy')[()]
```

```python
conditions = data.keys()
for condition in conditions:
    clones = data[condition]
    print('%24s : %d CLONEs' % (condition, len(clones)))
```

```python
for feature in data['AurA +Tpx2 +phos'][0].keys():
    print(feature)
```

```python
def hexbin_plot(x_key, y_key, prefix):
    sns.set(style="white", color_codes=True);
    conditions = list(data.keys())
    for (plot_index, condition) in enumerate(conditions):
        clones = data[condition]
        nclones = len(clones)
        # Concatenate data
        x_data = np.concatenate([clone[x_key] for clone in clones])
        y_data = np.concatenate([clone[y_key] for clone in clones])
        # Plot
        fig = sns.jointplot(x_data, y_data, kind="hex", color="#4CB391", stat_func=None, gridsize=50).set_axis_labels(x_key, y_key);
        fig.savefig(prefix + condition + '.png', dpi=300);
```

```python
y_key = 'S283 O - R286 H distance (A)'
x_key = 'L225 CA - S284 CA distance (A)'
hexbin_plot(x_key, y_key, 'S283 O - R286 H + L225 CA - S284 CA : ')
```

```python
y_key = 'P282 O - R285 H distance (A)'
x_key = 'L225 CA - S284 CA distance (A)'
hexbin_plot(x_key, y_key, 'P282 O - R285 H + L225 CA - S284 CA : ')
```
