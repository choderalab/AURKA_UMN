```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
plt.rc('font',family='serif')
import re
import pyemma.coordinates
import seaborn as sns
from glob import glob
sns.set_style("white")
%pylab inline
%matplotlib inline
```

```python
phos_data_tpx = glob('/Users/albaness/github/AURKA_UMN/data/levinson-coord-data/distances_AURKA_AURKA_phos_tpx2-*.npy')
nophos_data_tpx = glob('/Users/albaness/github/AURKA_UMN/data/levinson-coord-data/distances_AURKA_AURKA_nophos_tpx2-*.npy')
phos_data_notpx = glob('/Users/albaness/github/AURKA_UMN/data/levinson-coord-data/distances_AURKA_AURKA_phos_notpx2-*.npy')
nophos_data_notpx = glob('/Users/albaness/github/AURKA_UMN/data/levinson-coord-data/distances_AURKA_AURKA_nophos_notpx2-*.npy')
```

```python
data_phos_tpx = np.load(phos_data_tpx[0])
data_nophos_tpx = np.load(nophos_data_tpx[0])
data_phos_notpx = np.load(phos_data_notpx[0])
data_nophos_notpx = np.load(nophos_data_notpx[0])
```

```python
colors = ["amber","dark lavender","cerulean", "pale red"]
```

```python
with sns.xkcd_palette(colors):
    plt.figure(figsize=(20,15), dpi=300)
    labels = []
    sns.distplot(data_phos_tpx,bins=None, norm_hist=True)
    labels.append('+Phos +TPX')
    sns.distplot(data_nophos_tpx,bins=None, norm_hist=True)
    labels.append('-Phos +TPX')
    sns.distplot(data_phos_notpx,bins=None, norm_hist=True)
    labels.append('+Phos -TPX')
    sns.distplot(data_nophos_notpx,bins=None, norm_hist=True)
    labels.append('-Phos -TPX')
    plt.title('%s-%s' % ('T288', 'R255'), fontsize=50)
    plt.legend(labels, fontsize=45);
    plt.xlabel('T288 $\\alpha$C - R255 Cz Distance ($\\AA$)', fontsize=40);
    plt.ylabel('probability', fontsize=45)
    plt.yticks([], fontsize=30);
    plt.xticks(fontsize=40)
    plt.axvline(x=5, ymin=0, ymax=1, c='black', ls='--', linewidth=4)
    plt.xlim([2,14])
    plt.tight_layout()
plt.savefig('t288-r255cz-all4.pdf')
```
